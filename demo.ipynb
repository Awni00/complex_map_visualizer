# Input-Output Spaces

```python
from complex_vis import *

from sympy import *
```

```python
X_grid = generate_grid()
X_grid_55 = generate_grid(x_range=(-5, 5), y_range=(-5,5))

X_polar = generate_polar_grid(r_max=10, theta_res=500)
```

```python
# map of rotation by 45degrees
plot_map(X_grid_55, lambda z: z*exp(I*pi/4));
```

```python
# square map on [-5,5]x[-5,5]
plot_map(X_grid_55, lambda z: z**2);
```

```python
# cube map on [-5,5]x[-5,5]
plot_map(X_grid_55, lambda z: z**3);
```

```python
# exponential map on [0,1]x[0,1]
plot_map(X_grid, exp);
```

```python
# exponential map on [-5, 5]x[-5,5]
plot_map(X_grid_55, exp, color='#269df2');
```

```python
# exponential map on polar grid
plot_map(X_polar, exp);
```

```python
# inversion map on [-5,5]x[-5,5]
plot_map(X_grid_55, lambda z: 1/z);
```

```python
# inversion map on polar grid
plot_map(X_polar, lambda z: 1/z);
```

```python
# log map on [0,1]x[0,1]
plot_map(X_grid_55, log);
```

```python
# log map on polar grid
plot_map(X_polar, log);
```

```python
# define mobius transform corresponding to [[1+i, 0],[1, 0]]
mb_tfm1 = define_mobius_transform([[1+I, 0],[1, 1]])

# plot above mobius transform on [0,1]x[0,1]
plot_map(X_grid_55, mb_tfm1);
```

# Domain Colouring

```python
from complex_vis import *
```

```python
f = lambda z: (z**2 - 1)*(z - 1j)/(z**2 + 2j)
f_str = '$\\frac{(z^2 - 1)(z - i)}{z^2 + 2i}$'
```

```python
fig, ax = domain_color_plot(f, (-2, 2), (-2,2), mag_as='saturation', title=f_str, resolution=(2000,2000))
```
